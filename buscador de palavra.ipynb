#### Faz o download de todas as palestras do Pathwork e coloca numa pasta local

```python
pwBrasilURL = 'http://pathworkbrasil.com.br'
pwPalestras = pwBrasilURL+'/detalhes_home.php?id=6&titulo=Palestras%20dispon%C3%ADveis%20para%20uso%20pessoal'

# calls url and returns page text
def get_page(url):
    rsp=r.get(url)
    if rsp.status_code == 200:
        return rsp.text
    print("error: unexpected http status returned. Status code:", rsp.status_code)
    print("headers:", rsp.headers)
    print("response body:", rsp.text)
    return None

# load content of the website into a variable
page = get_page(pwPalestras)
# parse page using bs4
soup = bs(page, 'html.parser')

# cria uma lista com todos os links de todas as 258 palestras
pdfLinkList = list()

# procura os links de cada palestra e coloca na lista
for link in soup.find_all(href=re.compile("pw-\d{3}.pdf", re.IGNORECASE)):
    pdfLinkList.append(pwBrasilURL+link.get('href'))

# avise se nao tem o resultado de 258 palestras
if len(pdfLinkList) != 258:
    print("lista de palestras incorreto, deveria ser 258 mas tem apenas",len(pdfLinkList))
# else:
#     print("links de todas as",len(pdfLinkList),"palestras do Pathwork achados com sucesso!")

arquivosPalestras = str(pathlib.Path().absolute()) + '/palestras'
if not os.path.exists(arquivosPalestras):
    os.makedirs(arquivosPalestras)

for pwURL in pdfLinkList:
    f = arquivosPalestras+'/'+pwURL.split('/')[len(pwURL.split('/'))-1]
    if not os.path.exists(f):
        print("fazendo download da palestra:",pwURL.split('/')[len(pwURL.split('/'))-1])
        pdf = r.get(pwURL)
        pwURL.split()
        with open(f, 'wb') as arquivo:
            arquivo.write(pdf.content)
```

## Limpar, organizar e armazenar palestras em um dataframe
Dados armazenados em memoria, com rapido acesso.

```python
arquivosPalestras = str(pathlib.Path().absolute()) + '/palestras/'

palestras, textos = list(),list()
df = pd.DataFrame(data={"palestras": palestras, "textos": textos})
# c=0
for palestra in os.listdir(arquivosPalestras):
    if not palestra.endswith('.pdf'):
        continue
    # print('PALESTRA ===>>> ',palestra)
    texto = process_pdf(arquivosPalestras+palestra)
    textos.append(texto.replace('-\n',''))
    palestras.append(palestra.split('.')[0])
    # c+=1
    # if c > 3:
    #     break

df.palestras = palestras
df.textos = textos
```

#### Cria um novo dataframe com o codigo da palestra e o texto separado por cada paragrafo
Assim a pesquisa de palavras retorna um paragrafo e nao uma palestra

```python
todos_paragrafos, codigo_palestras = list(), list()
dt = pd.DataFrame(data={"codigo_palestras": codigo_palestras, 'paragrafos': todos_paragrafos})

for i in range(df.palestras.count()):
    for paragraph in df.textos[i].split(' \n \n'):
        if re.search('\w',paragraph):
            todos_paragrafos.append(re.sub('[\n\t]', '', paragraph.strip()))
            codigo_palestras.append(df.palestras[i])
dt.paragrafos = todos_paragrafos
dt.codigo_palestras = codigo_palestras
```

#### Machine learning test

```python
nlp = spacy.load("pt_core_news_sm")
tok_text=[] # OUTPUT for our tokenised corpus
text2 = dt.codigo_palestras.str.lower().values+' '+dt.paragrafos.str.lower().values

#Tokenising using SpaCy:
for doc in tqdm(nlp.pipe(text2, n_process=2, disable=["tagger", "parser","ner"])):
   tok = [t.text for t in doc ]#if t.is_alpha]#(t.is_ascii and not t.is_punct and not t.is_space)]
   tok_text.append(tok)
```

```python
from gensim.models.fasttext import FastText

ft_model = FastText(
    sg=1, # use skip-gram: usually gives better results
    vector_size=100, # embedding dimension (default)
    window=10, # window size: 10 tokens before and 10 tokens after to get wider context
    min_count=5, # only consider tokens with at least n occurrences in the corpus
    negative=15, # negative subsampling: bigger than default to sample negative examples more
    min_n=2, # min character n-gram
    max_n=5 # max character n-gram
)
ft_model.build_vocab(tok_text) # tok_text is our tokenized input text - a list of lists relating to docs and tokens respectivley

ft_model.train(
    tok_text,
    epochs=6,
    total_examples=ft_model.corpus_count, 
    total_words=ft_model.corpus_total_words)

ft_model.save('_fasttext.model') #save
ft_model = FastText.load('_fasttext.model') #load
```

```python
ft_model.wv.most_similar("compromisso", topn=10, restrict_vocab=5000)
```

```python
weighted_doc_vects = []

for i,doc in tqdm(enumerate(tok_text)):
  doc_vector = []
  for word in doc:
    # if word.isalpha():
    vector = ft_model.wv[word]
    weight = (bm25.idf[word] * ((bm25.k1 + 1.0)*bm25.doc_freqs[i][word])) / (bm25.k1 * (1.0 - bm25.b + bm25.b *(bm25.doc_len[i]/bm25.avgdl))+bm25.doc_freqs[i][word])
    weighted_vector = vector * weight
    doc_vector.append(weighted_vector)
  doc_vector_mean = np.mean(doc_vector,axis=0)
  weighted_doc_vects.append(doc_vector_mean)
  
#   pickle.dump( weighted_doc_vects, open( "weighted_doc_vects.p", "wb" ) ) #save the results to disc
```

```python
import nmslib

# create a matrix from our document vectors
data = np.vstack(weighted_doc_vects)

# initialize a new index, using a HNSW index on Cosine Similarity
index = nmslib.init(method='hnsw', space='cosinesimil')
index.addDataPointBatch(data)
index.createIndex({'post': 2}, print_progress=True)
```

```python
input = '196'.lower().split()

query = [ft_model.wv[vec] for vec in input]
query = np.mean(query,axis=0)

t0 = time.time()
ids, distances = index.knnQuery(query, k=10)
t1 = time.time()
print(f'Searched {df.shape[0]} records in {round(t1-t0,4) } seconds \n')
for i,j in zip(ids,distances):
  print(round(j,2))
  print(dt.paragrafos.values[i])
```
